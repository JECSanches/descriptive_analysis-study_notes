## 1 - Tipos de variáveis e análise univariada e bivariada <a name='tvaub'></a>

### A - Tipos de variáveis <a name='tvar'></a>

**Quantitativas**: (Representam quantidades, valores)
- **Discretas:** Têm seu domínio (conjunto de possíveis valores) enumerável. Comumente são contagens de algo. Ex: Nota em uma prova (caso não haja muitos valores intermediários), resultado de um dados, número de objetos vendidos, quantidade de filhos, etc.

- **Contínuas:** São quantidades que têm seu domínio (conjunto de possíveis valores) não enumerável. Em geral são medidas de um atributo físico ou subproduto de outras variáveis (médias, somas etc). Ex: Peso, medidas de tamanho, temperatura, salário (é discreto, mas pode ser tratado como contínuo devido ao número de possibilidades).

A variável contínua pode ser analisada em **histogramas**, em que o domínio da variável é "fatiado" em intervalos iguais e se tem uma contagem (de ocorrências) em cada intervalo; Em **boxplot**, em que a barra central da caixa indica a mediana, ou seja, metade dos dados é maior e a outra metade menor. A barra inferior indica o *primeiro quartil* e a barra superior indica o *terceiro quartil*. A distância entre o primeiro e o terceiro quartil é chamada de *intervalo interquartil*. As barras horizontais, superior e inferior, são obtidas com o *terceiro quartil + 1.5 x do intervalo interquartil* e o *primeiro quartil - 1.5 x do intervalo interquartil*, respectivamente. Valores que estão além de tais barras são potenciais pontos discrepantes, outliers, pontos muito distantes da média.

**Qualitativas**: 

São categorias, classificações ou qualidades de algo. Não faz sentido fazer contas com tais dados.
- **Ordinais:** Possuem uma "ordem" implícita. Ex: Tamanho - P, M ou G; Opinião - Concorda, neutro ou discorda; Classe social - A, B, C, D e E. Podem ser analisadas com gráficos de barras, analisando a frequência e sempre mantendo a ordenação.
  
- **Nominais:** Também são categorias, mas sem ordem implícita. Ex: Sabor - Limão, uva, chocolate; Estado civil - casado, solteiro, desquitado etc.; Tipo de jogo: plataforma, corrida, esporte etc. Podem ser analisadas com gráficos de barras, frequências e contagens.

### 3.1.3 - Gráfico de Pareto <a name='pareto'><a/>

O diagrama de Pareto é um gráfico de colunas que ordena as frequências das ocorrências, da maior para a menor, permitindo a priorização dos problemas, procurando levar a cabo o princípio de Pareto, isto é, há muitos problemas sem importância diante de outros mais graves (extraído de [wikipedia](https://pt.wikipedia.org/wiki/Diagrama_de_Pareto)).

Ele é baseado no princípio de Pareto, também conhecido como a regra 80/20, que sugere que aproximadamente 80% dos efeitos vêm de 20% das causas. Esse princípio pode ser aplicado em várias áreas, como controle de qualidade, gestão de projetos, economia e muitos outros campos.

````python
# Gráfico de Pareto
from matplotlib.ticker import PercentFormatter

# Gerando um DataFrame
df = pd.DataFrame({'contagem': [43, 20, 174, 5, 2, 1, 1, 0]})
df.index = ['Produto com defeito', 'Produto errado', 'Tempo de entrega',
           'Horário de entrega', 'Mau atendimento', 'Valor errado', 'Não serve',
           'Desistência']

df = df.sort_values(by='contagem', ascending=False)
df['pctacum'] = df['contagem'].cumsum()/df['contagem'].sum()*100
# .cumsum() - soma cumulativa

df
````

````python
# Elaborando o gráfico
fig, ax = plt.subplots()

# Primeiro eixo é a barra da contagem
ax.bar(df.index, df['contagem'])
# twinx() faz um gráfico como função do eixo secundário
ax2 = ax.twinx() # cria uma eixo gêmio, secundário
ax2.plot(df.index, df['pctacum'], color='r', marker='o')
ax2.yaxis.set_major_formatter(PercentFormatter())

ax.tick_params(axis='x', rotation=45)
ax2.tick_params(axis='y', colors='r')
plt.show()
````

As três primeiras reclamações somam mais de 95% do todo, ou seja, atendendo as reclamações mais relevantes (tempo de entrega, produto com defeito, produto errado), boa parte do problema da empresa será sanado. 

Isso destaca a relevância do gráfico de Pareto.

## 5 - Seaborn e análises bivariadas discretas <a name='seabornB'><a/>

Gráficos de barras com duas entradas

- `sns.countplot()` altura denota a frequência
- `sns.catplot()` pode-se determinar a componente vertical
- `df.crosstab()` Calcule uma tabulação cruzada simples de dois (ou mais) fatores.

````python
# As bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
````

````python
# Carregando o banco de dados de gorjetas (contido no sns)
tips = sns.load_dataset('tips')
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])
tips.head()
````

````python
# Distribuição de fumantes e não fumantes por dia
# Avaliando duas variáveis (['day','smoker'])

plt.rc('figure', figsize=(6,3))
sns.countplot(
    x = 'day',
    hue = 'smoker',
    data = tips[tips['tip_pct'] < 1],
)
# dataset filtrado

# Pode-se comparar a frequência de fumantes e não fumantes por dia
````

````python
# Invertendo a relação
# Pode-se ler agora o dia da semana em que há mais 
# ou menos fumantes

plt.rc('figure', figsize=(8,4))
sns.countplot(
    x = 'smoker',
    hue = 'day',
    data = tips[tips['tip_pct'] < 1],
)
````

````python
# Tabela cruzada (criando uma tabela de frequências)
freq = pd.crosstab(tips['smoker'], tips['day'])
freq

# O método .crosstab() cruza as duas variáveis
# Mesmos valores do gráfico imediatamente acima
````

````python
freq.sum(axis=1) # Soma as colunas
````

````python
# Obtendo a porcentagem de fumantes e 
# não fumantes em relação do dia da semana

freq_pct_linha = freq.div(freq.sum(axis=1), axis=0)
# Basicamente divide o valor total de "yes" e "no" pelas linhas
# da tabela "freq", obtendo assim a porcentagem para cada dia
# em relação do todo 
# Pode-se verificar o primeiro elemento com o "print" abaixo
print('--> O primeiro elemento da tabela será: {}'.
      format(freq.iloc[0,0]/freq.sum(axis=1).iloc[0]))

freq_pct_linha
````

````python
# Agora os valores estão normalizados, sua soma equivale a 1
freq_pct_linha.sum(axis=1)

# Normalizar: ajustar as variáveis em um intervalo de 0 a 1
````

````python
# Agora pode-se gerar um gráfico mais interessante
# Como está normalizado, a soma das barras é 1

freq_pct_linha.plot.bar()

# Tem-se agora uma correção de escala, tornando a comparação mais
# "correta"
````

````python
# Fazendo agora o inverso, ou seja, o dia da semana com
# maior porcetual de fumantes (basta inverter a análise)
freq = pd.crosstab(tips['day'], tips['smoker'])
freq

# Em geral deixa-se a variável de interesse na coluna (fumante, não fumante)
````

````python
# Obtendo as porcentagens
freq_pct_linha = freq.div(freq.sum(axis=1), axis=0)
print(freq_pct_linha.sum(axis=1)) # Normalizado
freq_pct_linha
````

````python
# Plot
freq_pct_linha.plot.bar()
# Método de plot do dataframe (Pandas)
````

### 5.1 - Função `catplot` do Seaborn <a name='funcCatplot'><a/>

Esta função fornece acesso a diversas funções em *axes-level* que mostram a relação entre uma variável numérica e uma ou mais variáveis ​​categóricas usando uma das diversas representações visuais. O parâmetro `kind` seleciona a função de *axes-level* subjacente a ser usada. [Documentação](https://seaborn.pydata.org/generated/seaborn.catplot.html)

Algumas opções do parâmetro `Kind`: “strip”, “swarm”, “box”, “violin”, “boxen”, “point”, “bar”, or “count”.

````python
# Avaliando o dia da semana em que o percentual de gorjetas é maior

sns.catplot(
    x = 'day',
    y = 'tip_pct',
    kind = 'bar',
    data = tips[tips['tip_pct'] < 1]
)

# Por padrão, tem-se a média no eixo y
````

````python
# Utilizando outro "kind"

sns.catplot(
    x = 'day',
    y = 'tip_pct',
    data = tips[tips['tip_pct'] < 1],
    kind = 'swarm'
)
````

````python
# Determinando o dia da semana em que fumantes e 
# não fumantes consomem mais

sns.catplot(
    x = 'day',
    y = 'total_bill',
    hue = 'smoker', # variável de interesse
    kind = 'bar', 
    estimator = sum, # Agora tem-se a soma e não a média
    data = tips[tips['tip_pct'] < 1],
    errorbar = None # O intervalo de confiança não faz sentido nesta análise
)
````

````python
# Analisando outro banco de dados do Seaborn ("penguins")
penguins = sns.load_dataset('penguins')
penguins.head()
````

### 6.1 - Histograma <a name='histo'><a/>

É possível utilizar os métodos do Pandas, matplotlib ou Seaborn. Todos trabalham na base do matplotlib, mas o Seaborn possui mais recursos para fornecer gráficos mais informativos e detalhados.

[Seaborn](https://seaborn.pydata.org)

````python
# Histograma 

sns.displot(
    data = penguins,
    x = 'body_mass_g',
    hue = 'sex',
    bins = 20,
    element = 'step',
    alpha = .3,
    kde = True
)
````

É interessante notar que o histograma "female" é modal, possui duas regiões diferentes de máximos locais da frequência da massa do pinguim. Desta forma, é possível que haja um outra variável explicando a variabilidade da massa dos pinguins.

- Analisando no boxplot para identificar a variável...

### 6.2 - Boxplot <a name='boxplot'><a/>

Novamente, é mais interessante utilizar também o Seaborn para o método `boxplot`.

````python
plt.rc('figure', figsize=(6,5))
sns.boxplot(x='sex', y='body_mass_g', data=penguins)
````

````python
# Utilizando agora "sex" como variável de interesse 
sns.boxplot(
    data=penguins, 
    x='species', 
    y='body_mass_g', 
    hue='sex', 
    orient='x' # Orientação do plot (vertical, horizontal)
)
````

Com base neste painel, pode-se afirmar que o comportamento que é visto no histograma com dois máximos locais se deve a "specie" Gentoo, pois é significativamente mais pesada que os demais, justificando assim o comportamento do histograma.

### 6.3 - Scatterplot <a name='scatter'><a/>

O Seaborn tem diversos recursos para fazer um gráfico de dispersão com uma e diversas variáveis.

- `scatterplot`, gera gráficos de dispersão simples
- `jointplot` tem recursos mais diversificados e fornecem gráficos bem interessantes
- `pairgrid` monta uma matriz de gráficos de dispersão com recursos interessantes para visualização da variabilidade dos dados.

````python
# Usando transparência

plt.rc('figure', figsize=(7,5))
sns.scatterplot(
    x = 'bill_length_mm',
    y = 'bill_depth_mm',
    data = penguins,
    alpha = .5
)
````

````python
# Adicionando parâmetros

plt.rc('figure', figsize=(8,5))
sns.scatterplot(
    data = penguins,
    x = 'bill_length_mm',
    y = 'bill_depth_mm',
    hue = 'species', # variável de interesse
    size = 'body_mass_g', # tamanho do círculo em relação a massa
    alpha = .5
)
````

### 6.4 - Gráficos com muitos pontos <a name='grafMP'><a/>

Quando se tem muitos pontos, torna-se complicado encontrar padrões. Algumas ideias são:

- Amostrar os dados
- Utilizar transparência (alpha menor que 1)
- Fazer o `scatterplot 2D`

#### 6.4.1 - Histograma 2D <a name='hist2D'><a/>


````python
sns.jointplot(
    x = 'bill_length_mm',
    y = 'bill_depth_mm',
    data = penguins, 
    kind = 'hist',
    bins = [15, 15] # Reflete o número de pontos da grade
)
````

````python
sns.jointplot(
    x = 'bill_length_mm',
    y = 'bill_depth_mm',
    data = penguins, 
    hue = 'species', # Variável de interesse
    alpha = .6
)
````

Quando temos muitos pontos o scatterplot dificilmente apresenta uma visualização prática. Para isso podem ser utilizados recursos como transparência e histogramas 2D.

### 6.4 - Matrizes de dispersão <a name='MD'><a/>

````python
# Matrizes de dispersão
sns.pairplot(penguins, hue='species')

# As variáveis quantitativas dão origem a um gráfico.
````

````python
# Utilizando outros recursos do pairgrid
g = sns.PairGrid(penguins, hue='species')
g.map_upper(sns.histplot)

# Parte superior da matriz scatterplot
g.map_upper(sns.scatterplot) 

# Parte inferior da matriz kdeplot (curvas de nível)
g.map_lower(sns.kdeplot, fill=True)

# Na diagonal histograma com o kde (suavização)
g.map_diag(sns.histplot, kde=True)

# Adiciona legenda e ajusta
g.add_legend(title=" species ", adjust_subtitles=True)
````

````python
# Utilizando outros recursos do pairgrid
g = sns.PairGrid(penguins, hue='species')
# g.map_upper(sns.histplot)

# Parte superior da matriz scatterplot
g.map_upper(sns.scatterplot) 

# Parte inferior da matriz kdeplot (curvas de nível)
g.map_lower(sns.kdeplot, fill=True)

# Na diagonal histograma com o kde (suavização)
g.map_diag(sns.histplot, kde=True)

# Adiciona legenda e ajusta
g.add_legend(title=" species ", adjust_subtitles=True)
````
